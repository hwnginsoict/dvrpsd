```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Pivot the data to format it for the heatmap
heatmap_data = cp_200_merge.pivot(index="ited", columns="sized", values="obj")

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.set(font="Times New Roman")  # Set default font for seaborn
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".6f",
    cmap="YlGnBu_r",
    annot_kws={"fontname": "Times New Roman", "fontsize": 15},  # Set font for numbers
    cbar_kws={'label': 'obj Value'},  # Set font for color bar label
)

# Adjust the title, x, and y labels font properties
# plt.title("Dataset 200", fontsize=19, fontname="Times New Roman", fontweight='bold')
plt.xlabel("sizemax", fontsize=16, fontname="Times New Roman",  fontweight='bold')
plt.ylabel("itermax", fontsize=16, fontname="Times New Roman",  fontweight='bold')

# Customize color bar label font
cbar = plt.gca().collections[0].colorbar
cbar.set_label("objective", fontsize=16, fontname="Times New Roman", fontweight='bold')

min_val, max_val = round(heatmap_data.min().min(), 3), round(heatmap_data.max().max(), 3)
cbar.set_ticks(np.linspace(min_val, max_val, 5))  # Set 5 evenly spaced ticks

plt.xticks(fontsize=14, fontname="Times New Roman",  fontweight='bold')  # Increase font size for x-axis tick labels
plt.yticks(fontsize=14, fontname="Times New Roman",  fontweight='bold')  # Increase font size for y-axis tick labels


plt.savefig("CP_200.pdf", format='pdf', dpi=300, bbox_inches='tight')
# Show plot
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('F:\CodingEnvironment\dvrpsd\experiment\EP200.xlsx')
```

```python
# Initialize dictionaries to store results
count_dict = {f' fheu{i+1}': {} for i in range(6)}
improvement_percentage_dict = {f' fheu{i+1}': {} for i in range(6)}
total_improvement = {f' fheu{i+1}': 0 for i in range(6)}

# Group by 'Type' and calculate counts and improvements
for t, group in df.groupby('Type'):
    for i in range(1, 7):
        fheu_col = f' fheu{i}'
        
        # Count the times DACO is smaller than fheu
        count = (group['DACO'] < group[fheu_col]).sum()
        count_dict[fheu_col][t] = count
        
        # Calculate improvement percentage for all rows
        improvements = group[fheu_col]
        improvement_percentage = ((improvements - group['DACO']) / improvements).mean() * 100
        
        # Store improvement percentage per type and fheu
        improvement_percentage_dict[fheu_col][t] = improvement_percentage
        
        # Accumulate total improvement for overall average calculation
        total_improvement[fheu_col] += improvement_percentage

# Calculate overall average improvement percentage for each fheu column
average_improvement_percentage = {fheu: total_improvement[fheu] / len(df['Type'].unique())
                                  for fheu in total_improvement}

# Display results
print("Counts where DACO is smaller than each fheu by Type:")
for fheu, counts in count_dict.items():
    print(f"{fheu}: {counts}")

print("\nImprovement Percentage for each fheu by Type:")
for fheu, improvements in improvement_percentage_dict.items():
    print(f"{fheu}: {improvements}")

print("\nOverall Average Improvement Percentage for each fheu:")
for fheu, avg_improvement in average_improvement_percentage.items():
    print(f"{fheu}: {avg_improvement:.2f}%")
```
